# Medical NER Project - Notebook 01
## Dataset Download and Preparation

This notebook handles the download and extraction of the MACCROBAT2018 dataset, which is used for training and evaluating the biomedical named entity recognition (NER) model.


## Step 1: Setup

Importing necessary libraries and defining constants for download location and dataset path.


```python
import os
import requests
import zipfile
```

## Step 2: Download the MACCROBAT2018 Dataset

The dataset is hosted on Figshare and downloaded as a ZIP file. We define a utility function to handle the download using `requests`.


```python
MACCROBAT_DATA_URL = "https://figshare.com/ndownloader/files/17493650"
```

```python
def download_zip_from_url(url, save_path, chunk_size=128):
    r = requests.get(url, stream=True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=chunk_size):
            fd.write(chunk)
```

```python
data_zipfile = "../data/MACCROBAT2018.zip"
MACCROBAT_data_dir = "../data/MACCROBAT"
```

```python
if not os.path.exists(MACCROBAT_data_dir):
    os.makedirs(MACCROBAT_data_dir)
    print(f"Directory '{MACCROBAT_data_dir}' created.")
else:
    print(f"Directory '{MACCROBAT_data_dir}' already exists.")
```

```python
download_zip_from_url(MACCROBAT_DATA_URL, data_zipfile)
```

## Step 3: Extract the Dataset

Once downloaded, the dataset ZIP file is extracted into the defined data directory.


```python
with zipfile.ZipFile(data_zipfile, 'r') as zip_ref:
    zip_ref.extractall(MACCROBAT_data_dir)
```

## Step 4: Post-processing and Validation

After extraction, the original ZIP file is deleted to save space. Finally, the number of `.txt` files that have corresponding `.ann` annotation files is counted to verify dataset integrity.


```python
os.remove(data_zipfile)
```

```python
print(f"Number of txt files corresponding with ann files is\
 {len([file for file in os.listdir(MACCROBAT_data_dir) if file.endswith('.txt') and file.split('.')[0]+ '.ann' in os.listdir(MACCROBAT_data_dir)])}")
```
